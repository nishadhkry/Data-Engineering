```python
%sql
--creating employee table
create table if not exists employee_table (
  emp_id int primary key,
  emp_f_name string,
  emp_l_name string,
  salaray int,
  hire_date date, 
  dept_id int
);
```

```python
%sql
--department table
create table if not exists department2(
  dept_id int,
  dept_name string
);
   
```

```python
%sql
--Departments with more than 3 employees using join
SELECT d.dept_id,d.dept_name,COUNT(e.emp_id) AS total_employees
FROM employee_table e
INNER JOIN department2 d
ON d.dept_id = e.dept_id
GROUP BY d.dept_id, d.dept_name
HAVING total_employees >=3;
```

```python
%sql
SELECT e.emp_id,e.emp_f_name,e.emp_l_name,d.dept_name
FROM employee_table e
INNER JOIN department2 d
ON d.dept_id = e.dept_id
WHERE e.dept_id IN (
    SELECT dept_id
    FROM employee_table
    GROUP BY dept_id
    HAVING COUNT(*) >= 3);
```

```python
%sql
SELECT d.dept_id, d.dept_name
FROM department2 d
LEFT JOIN employee_table e
ON d.dept_id = e.dept_id
WHERE e.emp_id IS NULL;
```

```python
%sql
-- Find employees without department using JOIN

SELECT e.emp_id, e.emp_f_name,e.emp_l_name
FROM employee_table e
LEFT JOIN department2 d
ON e.dept_id = d.dept_id
WHERE d.dept_id IS NULL;
```

```python
%sql
--Find second highest salary
SELECT salaray
FROM (
    SELECT salaray,
           DENSE_RANK() OVER (ORDER BY salaray DESC) AS salary_rank
    FROM employee_table
)
WHERE salary_rank = 2
GROUP BY salaray;
```

```python
%sql
-- Count employees hired per year

SELECT YEAR(hire_date) AS hire_year,
       COUNT(*) AS total_employees
FROM employee_table
GROUP BY YEAR(hire_date)
ORDER BY hire_year;
```

```python

%sql
-- Calculate running total of salary(Window Function)
select emp_id,emp_f_name,emp_l_name,salaray ,
sum(salaray)over(order by emp_id) as running_total
from employee_table;
```
